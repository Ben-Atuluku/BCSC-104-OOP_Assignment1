```python
plt.figure(figsize=(20, 10) )
plt.title("Attack Frequency Per Year")
sb.countplot(
             data = dt,
             y = 'Year',
             color = sb.color_palette()[2],
             order = year_order
            )
plt.xticks(rotation = 45)
```

```python
plt.figure(figsize=(20, 10) )
plt.title("Top 20 Locations of Attack")
sb.countplot(
             data = dt,
             y = 'Location',
             color = sb.color_palette()[2],
             order = location_order[:20]
            )
plt.xticks(rotation = (0))
```

```python
plt.figure(figsize=(20, 10) )
plt.title("Motives For Attacks")
sb.countplot(
             data = dt,
             y = 'Motive',
             color = sb.color_palette()[2],
             order = motive_order
            )
plt.xticks(rotation = (45))
```

```python
plt.figure(figsize=(20, 10) )
plt.title("Attacked Industry")
sb.countplot(
             data = dt,
             y = 'Industry',
             color = sb.color_palette()[2],
             order = industry_order
            )
plt.xticks(rotation = (45))
```
